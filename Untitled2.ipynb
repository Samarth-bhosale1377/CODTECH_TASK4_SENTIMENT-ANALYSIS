```python
import pandas as pd
import numpy as np
import nltk
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
data = {
    'Review': [
        "I love this product! It works great.",
        "Worst experience ever, not recommended.",
        "Average quality, nothing special.",
        "Totally worth the money! Great quality.",
        "Poor packaging, arrived damaged."
    ]
}
df = pd.DataFrame(data)
df.head()
```

```python
# Convert to lowercase
df['clean_review'] = df['Review'].str.lower()

# Remove punctuation and unwanted characters
df['clean_review'] = df['clean_review'].str.replace('[^a-zA-Z ]', '', regex=True)

df.head()
```

```python
# Get polarity and subjectivity scores
df['polarity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

# Classify sentiment
def get_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['polarity'].apply(get_sentiment)

df.head(10)
```

```python
sentiment_counts = df['sentiment'].value_counts()

plt.figure(figsize=(6,4))
sentiment_counts.plot(kind='bar', color=['green','red','gray'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
text = " ".join(df['clean_review'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Reviews")
plt.show()
```
